# Creating a Filter, Edge Detection

### Import resources and display image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Read in the image
image = cv2.imread('images/city_hall.jpg')

# Make a copy of the image
image_copy = np.copy(image)

# Change color to RGB (from BGR)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

### Covert the image to grayscale

```python
# Convert to grayscale for filtering
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap = 'gray')
```

### Create a custom kernel

```python
# 3x3 array for edge detection
# this is to only detect vertical edges
# Sobel Filter

sobel_x = np.array([[ -1, 0, 1],
                  [-2, 0, 2],
                  [-1, 0, 1]])

# Perform convolution using filter2D, which has inputs: (grayscale image, bit-dept, kernel)
filtered_image = cv2.filter2D(gray, -1, sobel_x) # (input image, type of output image (-1 means same as the input image), kernel)

plt.imshow(filtered_image, cmap='gray')
```

### Create binary image

```python
# Create a threshold that sets all the filtered pixels to white
# Above a certain threshold

retval, binary_image = cv2.threshold(filtered_image, 100, 255, cv2.THRESH_BINARY) #(image on which to apply threshold, a low value to pixel value to pass through, pixel value to turn this all into, type od threshold)

plt.imshow(binary_image, cmap = 'gray')
```
