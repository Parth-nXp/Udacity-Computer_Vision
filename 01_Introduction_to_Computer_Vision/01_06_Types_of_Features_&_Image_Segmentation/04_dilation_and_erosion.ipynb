```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Reads in a image
image = cv2.imread('images/j.png', 0)

# Create a 5x5 kernel of ones
kernel = np.ones((10,10), np.uint8)

# Dilate the image
dilation = cv2.dilate(image, kernel, iterations = 1)
# 1st argument: an original binary image
# 2nd argument: a kernel that determines the size of the dilation (None will result in a default size)
# 3rd argument: a number of iterations to perform the dilation (typically = 1)

# a 10x10 kernel of ones, which move over an image, like a filter, and turn a pixel white if any of its surrounding pixels are white in a 10x10 window!

# Erode the image
erosion = cv2.erode(image, kernel, iterations = 1)


f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(20,10))

ax1.set_title('erosion')
ax1.imshow(erosion, cmap = 'gray')

ax2.set_title('original')
ax2.imshow(image, cmap = 'gray')

ax3.set_title('dilation')
ax3.imshow(dilation, cmap = 'gray')
```

### Opening

```python
# Read the image 
image = cv2.imread('images/j1.png', 0)

opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel=np.ones((15,15), np.uint8))

f, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))

ax1.set_title('original')
ax1.imshow(image, cmap = 'gray')

ax2.set_title('after opening')
ax2.imshow(opening, cmap = 'gray')
```

### Closing

```python
# Read the image 
image = cv2.imread('images/j2.png', 0)

closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel=np.ones((15,15), np.uint8))

f, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))

ax1.set_title('original')
ax1.imshow(image, cmap = 'gray')

ax2.set_title('after closing')
ax2.imshow(closing, cmap = 'gray')
```
