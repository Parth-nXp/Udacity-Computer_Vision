# Image Contouring

### Image resources and display image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

# Read in the image
image = cv2.imread('images/raised_hand.png')

# Make a copy of the image
image_copy = np.copy(image)

# Change color to RGB (from BGR)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

### Produce a binary image for finding contours

```python
# Convert to grayscale
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

# Create a binary threshold image
# inverted binary to show background black instead of showing it white
retval, binary = cv2.threshold(gray, 225, 255, cv2.THRESH_BINARY_INV)

plt.imshow(binary, cmap = 'gray')
```

### Find and draw the contours

```python
# Find contours from threshold image
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# first argument is the binary image
# contour retruval mode which is Tree
# countour apploximation method which is choosen as simple chain
# output is set of contours and list of hierarchy
# hierarchy is necessary if you have many contours nested within one another
# hierarchy defines their relation to one another

# Draw all contours on a copy of the original image
image_copy2 = np.copy(image_copy)
all_contours = cv2.drawContours(image_copy2, contours, -1, (0,255,0), 2)
# first argument for drawing the contour is the image on which contour needs to be drawn
# second argument is the set of all the contours that are needed to be drawn 
# third argument -1 means all of the contours
# fourt argument tells about the green color
# fifth argument defines the size of the contour that is 2 pt

plt.imshow(all_contours)
```
